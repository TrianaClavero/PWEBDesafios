# Aplicación con formato 

Desde el return en un endpoint podemos retornar directamente etiquetas HTML en lugar de texto plano o el JSON 

```python
from flask import Flask
from datetime import date, datetime

app = Flask(__name__)


@app.route('/edad/<fecha>')
def edad(fecha=None):
    try:
        fecha_actual = date.today()
        fecha_conv = datetime.strptime(fecha, "%Y-%m-%d") #Convertir texto a fecha 
        edad = fecha_actual - fecha_conv.date() #el resultado es en dias con horas
        edad = edad.days #obtengo solo los dias
        edad_anios = edad // 365.25 #el 365.25 toma en cuenta años biciesto para mas precision
        return f'<h1>La edad es: {int(edad_anios)}</h1>'
    except ValueError:
        return "<h1>Error: Formato inválido, usá YYYY-MM-DD</h1>"
    

app.run(use_reloader=False)
```

# Escape HTML

La función escape se utiliza para escapar caracteres especiales en HTML y así prevenir ataques de inyección, evitando que se interpreten como codigo HTML

```python
from flask import Flask
from markupsafe import escape

app = Flask(__name__)

@app.route('/<nombre>/<apellido>/<int:edad>')
def saludar(nombre, apellido, edad):
    return f'Hola {escape(nombre)} {escape(apellido)}, tenes {edad} anios'
```
