## ¿Qué es Flask?
Es un micro framework para Python que se utiliza para desarrollar aplicaciones web de manera simple y rápida. Para poder trabajar con Flask, primero necesitamos preparar un entorno de desarrollo en Python con el fin de organizar nuestro proyecto y evitar conflictos entre librerías.

## Preparación del entorno de programación


```python
# 1. Verificar instalación de Python
# Para saber si Python está instalado, abrimos la consola o terminal del sistema y ejecutamos el siguiente comando:
!python --version

# 2. Crear un directorio
# Debemos crear una carpeta para nuestro proyecto, donde vamos a guardar todos los archivos.
# Se puede hacer en Visual Studio Code o directamente en la termina/consola del sistema.
!mkdir proyecto_flask
# 3. Entrar al directorio
# Después ingresamos a esa carpeta para trabajar dentro de ella.
%cd proyecto_flask

# 4. Crear entorno virtual
# Creamos un entorno virtual dentro de la carpeta
!python -m venv venv
# El segundo “venv” es el nombre de la carpeta por lo que se puede elegir que nombre asignarle.

# 5. Activar entorno virtual
# Ahora activamos el entorno virtual que creamos:
#Windows:
!venv\Scripts\activate             
#Linux / macOS:
!venv/bin/activate                  
# 6. Instalar Flask
# Una vez activado el entorno virtual ya podemos instalar las librerías, en este caso Flask usando el comando “pip”
%pip install flask

# 7. Verificar instalación
# Luego verificamos que Flask se haya instalado correctamente mostrando la lista de librerías instaladas en el entorno
%pip list
```

## Primera aplicación

```python
# 8. Crear archivo principal (app.py)
%%file app.py
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return '<h1>¡Hola Mundo!</h1><p>Esta es nuestra primera app con Flask.</p>'

if __name__ == "__main__":
    app.run(debug=False, use_reloader=False)
```

# Ejecucion del servidor

```python
# Le indicamos al servidor qué aplicación tiene que lanzar con:
%env FLASK_APP=app.py
# Y acto siguiente ejecutamos el servidor con:
!flask run
!python app.py
```

# Aplicación con formato 

Desde el return en un endpoint podemos retornar directamente etiquetas HTML en lugar de texto plano o el JSON 

```python
# Creamos un segundo archivo para probar lógica y formato HTML estilizado
%%file app_formato.py
from flask import Flask
from datetime import date, datetime

app = Flask(__name__)


@app.route('/edad/<fecha>')
def edad(fecha=None):
    try:
        fecha_actual = date.today()
        fecha_conv = datetime.strptime(fecha, "%Y-%m-%d") #Convertir texto a fecha 
        edad = fecha_actual - fecha_conv.date() #el resultado es en dias con horas
        edad = edad.days #obtengo solo los dias
        edad_anios = edad // 365.25 #el 365.25 toma en cuenta años biciesto para mas precision
        return f'''
        <div style="font-family: Arial, sans-serif; border: 2px solid #3498db; padding: 20px; border-radius: 10px; width: fit-content;">
            <h1 style="color: #2980b9;">Resultado del Cálculo</h1>
            <p style="font-size: 1.2em;">La edad para la fecha <strong>{fecha}</strong> es: <span style="color: #e74c3c;">{int(edad_anios)} años</span>.</p>
        </div>
        '''
    except ValueError:
        return "<h1 style='color: #c0392b;'>Error: Formato de fecha inválido. Usá YYYY-MM-DD</h1>"
    

app.run(use_reloader=False)
```

# Escape HTML

La función escape se utiliza para escapar caracteres especiales en HTML y así prevenir ataques de inyección, evitando que se interpreten como codigo HTML

```python
%%file app_segura.py
from flask import Flask
from markupsafe import escape

app = Flask(__name__)

@app.route('/<nombre>/<apellido>/<int:edad>')
def saludar(nombre, apellido, edad):
    return f'Hola {escape(nombre)} {escape(apellido)}, tenes {edad} anios'
```

## Enrutamiento
El enrutamiento vincula una URL específica con una función de Python que genera la respuesta. Se define mediante el decorador `@app.route()`  \
1. Rutas estaticas  \
Son direcciones fijas. Si la URL no coincide exactamente, Flask devuelve un error 404.



```python
%%file app_rutas.py
@app.route('/')
def index():
    return 'Inicio'

@app.route('/contacto')
def contacto():
    return 'Página de contacto'
```

2. Rutas Dinámicas  \
Permiten capturar valores de la URL usando la sintaxis <nombre_variable>. El valor capturado se pasa como argumento a la función.

```python
%%file app_rutas.py
@app.route('/saludo/<nombre>')
def saludar(nombre):
    return f'Hola, {nombre}!'
```
